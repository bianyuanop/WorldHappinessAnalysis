```python
import pandas as pd
import altair as alt

alt.data_transformers.disable_max_rows()
```

```python
sgd = pd.read_csv('./dataset/world_happiness/sustainable_develop_goals/sdg-csv-zip-7-mb-/SDGData.csv')
sgd = sgd.iloc[:, :sgd.shape[1]-1]
sgd['index'] = sgd.reset_index().index
sgd.head()
```

```python
print(sgd.shape)
len(sgd['Indicator Code'].unique())
print(sgd.columns)
```

```python
sgd_long = pd.wide_to_long(sgd, '', j='year', i='index', suffix='\d*').reset_index()
sgd_long.drop('index', axis=1)
```

```python
print(sgd_long.shape)
sgd_long.head()
```

```python
sgd_long = sgd_long.rename(columns={'': 'value'})
print(sgd_long.columns)
sgd_long.drop('index', axis=1)
```

```python
sgd_long = sgd_long.dropna(axis=0)
```

```python
indicators = sgd_long['Indicator Code'].unique()
indicator_names = sgd_long['Indicator Name'].unique()

len(indicator_names) == len(indicators)
```

```python

for name in indicator_names:
    if 'rural' in name or 'urban' in name:
        print(name)
```

```python
indicators_count = {indicator: sgd_long[sgd_long['Indicator Code'] == indicator]['value'].count() for indicator in indicators}
```

```python
essential_indicator = max(indicators_count, key=indicators_count.get)
```

```python
sgd_essential = sgd_long[sgd_long['Indicator Code'] == essential_indicator]
sgd_essential
```

```python
alt.Chart(sgd_essential).encode(
    x='year:N',
    y='value',
    color='Country Code'
).mark_line()
```
